```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
print(x)
```

```python
y=3*x
print(y)
```

```python
plt.plot(x,y)
plt.show
```

```python
#adding label
plt.plot(x,y,color="b",linestyle="-.",linewidth=2)
plt.title("Line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show
```

```python
#adding two lines in same plot:
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle='-.',linewidth=2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.title("linePlot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle='-.',linewidth=2)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.grid(True)
plt.show
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle='-.',linewidth=2)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.grid(True)
plt.show
```

```python
#pai chart:-
import numpy as np
from matplotlib import pyplot as plt
x=["kamla","anil","vimal","rina","shan","Any","raj","Harry"]
y=[37,53,30,10,81,60,25,5]
plt.pie(y,labels=x)
plt.show()
```

```python
#doughnut-chart:-
plt.pie(y,labels=x,radius=1.5,autopct='%0.1f%%')
plt.pie([1],colors=['w'],radius=.5)
plt.show()
```
